```python
import pandas as pd
import numpy as np
import sklearn as sp
data=pd.read_csv("train.csv")
data_t=pd.read_csv("test.csv")
categorical_cols=data.select_dtypes(include='object').columns.tolist()[:-1]
numerical_cols=data.select_dtypes(include='number').columns.tolist()[1:]
from sklearn.preprocessing import OneHotEncoder
raw=data.fillna("unknown")
encoding=OneHotEncoder(sparse_output=False,handle_unknown="ignore")
encoding.fit(data[categorical_cols])
encoded_cols=list(encoding.get_feature_names_out(categorical_cols))
data[encoded_cols]=encoding.transform(raw[categorical_cols])
data_t[encoded_cols]=encoding.transform(data_t[categorical_cols])
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data[numerical_cols])
data[numerical_cols]=scaler.transform(data[numerical_cols])
data_t[numerical_cols]=scaler.transform(data_t[numerical_cols])
input=data[encoded_cols+numerical_cols]
input_t=data_t[encoded_cols+numerical_cols]
output=data.Personality
input.shape
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
model=HistGradientBoostingClassifier(max_iter=15000,l2_regularization=25)
model.fit(input,output)
model.predict(input_t)
```

```python
sample=pd.read_csv("sample_submission.csv")
sample.Personality=model.predict(input_t)
sample.to_csv("submission21.csv",index=False)
sample
```
